```python
#Codes for Energy plots (Figure 3)
# For \theta_0= 0.1 (wide cone)


import numpy as np
import matplotlib.pyplot as plt
import math

# Define constants
M = 0.1056583755 #GeV
E= 3.1 #GeV
y = 29.3
theta_0 = 0.1

# Redefining Constants-
b = math.sqrt(y**2 - 1)/y
a= M**2/(2*E)

#Check if values are working 
print("The value of b is:", b)
print("The value of a is:", a)

# Define theta values for plotting
theta = np.linspace(0, 0.2, 1000)

# Define the three equations
E_1e = a/(1 - b*np.cos(theta + theta_0))
E_2e = a/(1 - b*np.cos(theta - theta_0))
E_e = a/(1 - b*np.cos(theta))

# Plot the three equations
plt.plot(theta, E_1e, 'b', label='E1e', linewidth=2)
plt.plot(theta, E_2e, 'r', label='E2e', linewidth=2)
plt.plot(theta, E_e, 'k', label='Ee', linewidth=2)

# Add legend and labels
plt.legend()
plt.xlabel(r'$\theta$')
plt.ylabel('E')

# Adjust axis limits
plt.xlim(0, 0.2)
plt.ylim(0, 3.2)


# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import matplotlib.ticker as mtick

# Define constants
M = 0.1056583755 #GeV
E = 3.1 #GeV
y = 29.3
theta_0 = 0.1

# Redefining Constants-
b = math.sqrt(y**2 - 1)/y
a = M**2/(2*E)

# Check if values are working 
print("The value of b is:", b)
print("The value of a is:", a)

# Define theta values for plotting
theta = np.linspace(0, 0.2, 1000)

# Define the three equations
E_1e = a/(1 - b*np.cos(theta + theta_0))
E_2e = a/(1 - b*np.cos(theta - theta_0))
E_e = a/(1 - b*np.cos(theta))

# Plot the three equations
plt.plot(np.degrees(theta), E_1e, 'b', label='$E_{4_1}$', linewidth=2)
plt.plot(np.degrees(theta), E_2e, 'r', label='$E_{4_2}$', linewidth=2)
plt.plot(np.degrees(theta), E_e, 'k', label='$E_{4\;max}$', linewidth=2)

# Add legend and labels
plt.legend()
plt.xlabel(r'$\theta$ (degrees)')
plt.ylabel('$E_4$(GeV)')

# Adjust axis limits
plt.xlim(0, 11.5)
plt.ylim(0, 3.2)

# Add minor ticks
ax = plt.gca()
ax.xaxis.set_minor_locator(mtick.AutoMinorLocator())

# Add theta_0 value to the plot
plt.text(3, 3, r'$\theta_0 =$' + str(theta_0), fontsize=12)
# Show the plot
plt.show()
```
