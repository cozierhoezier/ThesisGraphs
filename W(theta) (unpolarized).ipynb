```python
import math
import matplotlib.pyplot as plt
import numpy as np

gamma = 29.3
beta = math.sqrt(gamma**2 - 1)/gamma
theta_0 = 0.01
y2=gamma**2
def a(theta):
    return 1 - beta*math.cos(theta)*math.cos(theta_0)

def b(theta):
    return beta*math.sin(theta)*math.sin(theta_0)

def Vparallel(theta):
    numerator = a(theta)*(1 + 1/(3*beta))
    denominator = (a(theta)**2 - b(theta)**2)**(3/2)
    second_term = (2*a(theta)**2 + b(theta)**2)/(6*y2*beta*((a(theta)**2 - b(theta)**2)**(5/2)))
    return (1/(4*math.pi*gamma**2))*(numerator/denominator - second_term)

def Vradial(theta):
    numerator = a(theta)*(1 - math.cos(theta_0)/(3*gamma*beta*math.sin(theta_0)))
    denominator = (a(theta)**2 - b(theta)**2)**(3/2)
    second_term = (2*a(theta)**2 + b(theta)**2)*(math.cos(theta_0) - beta*math.cos(theta))/(6*beta*gamma*math.sin(theta_0)*(a(theta)**2-b(theta)**2)**(5/2))
    return 1/(4*math.pi*gamma**2)*(numerator/denominator + second_term)

def zpolarized(theta):
    numerator = 1 + (beta - math.cos(theta))/(3*(1 - beta*math.cos(theta)))
    denominator = (gamma*(1 - beta*math.cos(theta)))**2
    return 1/(4*math.pi)*numerator/denominator

def Vunpolarized(theta):
    numerator = a(theta)
    denominator = (a(theta)**2 - b(theta)**2)**(3/2)
    return 1/(4*math.pi*gamma**2)*numerator/denominator

def PWunpolarized(theta):
    denominator = (gamma*(1 - beta*math.cos(theta)))**2
    return 1/(4*math.pi*gamma**2*(1 - beta*math.cos(theta))**2)

# plot the functions
# Generate values for theta
theta_values = np.linspace(0,0.2, 500)
Vparallel_values = [Vparallel(theta) for theta in theta_values]
Vradial_values = [Vradial(theta) for theta in theta_values]
zpolarized_values = [zpolarized(theta) for theta in theta_values]
Vunpolarized_values = [Vunpolarized(theta) for theta in theta_values]
PWunpolarized_values = [PWunpolarized(theta) for theta in theta_values]

plt.plot(theta_values, Vparallel_values, color='red', linestyle='solid', label='Vparallel')
plt.plot(theta_values, Vradial_values, color='red', linestyle='dashed', label='Vradial')
plt.plot(theta_values, zpolarized_values, color='black', linestyle='solid', label='zpolarized')
plt.plot(theta_values, Vunpolarized_values, color='red', linestyle='dotted', label='Vunpolarized')
plt.plot(theta_values, PWunpolarized_values, color='black', linestyle='dotted', label='PWunpolarized')

plt.legend()
plt.xlabel('$\theta$')
plt.ylabel('W($\theta$)')
plt.show()
```
